```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.heritagetype.com/pages/free-vintage-illustrations"
```

```python
def get_page_contents(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        print(f'Successfully retrieved the page_content \nurl: {url}\n')
        return response.content
    print(f'No content was retrieved \nurl: {url}\n')
    return None
```

```python
content = get_page_contents(url)
soup = BeautifulSoup(content, 'html.parser')
```

```python
resource_div = soup.find(class_='shg-box-content')
link_elements = resource_div.find_all(class_='shg-btn shg-cse')
```

```python
link_list = []
for element in link_elements:
    link = element.get('href')
    link_list.append(link)
```

```python
def get_link_list(url_list: list[str]) -> list[str]:

    link_list = []
    for url in url_list:
        content = get_page_contents(url)
        soup = BeautifulSoup(content, 'html.parser')

        resource_div = soup.find(class_='shg-box-content')
        link_elements = resource_div.find_all(class_='shg-btn shg-cse')

        for element in link_elements:
            link = element.get('href')
            link_list.append(link)
    
    return link_list
```

```python
url_list = [
    'https://www.heritagetype.com/pages/free-vintage-illustrations',
    'https://www.heritagetype.com/pages/free-vintage-illustrations-page-2'
]
```

```python
total_link_list = get_link_list(url_list)
total_link_list
```

```python
len(total_link_list)
```

```python
str_to_avoid = [
    'https://www.kittl.com',
    'https://www.heritagetype.com/pages/free-vintage-illustrations'
]
```

```python
final_urls = [url for url in total_link_list if url not in str_to_avoid]
```

```python
matches = [
    string for string in total_link_list
    if any(url in string for url in str_to_avoid)
]
matches
```

```python
final_urls = [
    url for url in total_link_list
    if not any(avoid_str in url for avoid_str in str_to_avoid)
]
```

```python
len(final_urls)
```

```python
with open("urls.txt", "w") as file:
    for url in final_urls:
        file.write(url + "\n")

print("URLs have been saved to 'urls.txt'.")
```
